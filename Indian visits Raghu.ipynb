```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
data = pd.read_csv('Top Indian Places to Visit 2.csv')
```

```python
print(data.columns)
```

```python
# Displaying the first few rows of the dataset
print(data.head(5))
```

```python
# Function to create a pie chart
def plot_pie_chart(data, column):
    plt.figure(figsize=(8, 8))
    plt.pie(data[column].value_counts(), labels=data[column].value_counts().index, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.title('Pie Chart')
    plt.show()
```

```python
# Function to create a scatter plot
def plot_scatter_plot(data, x_column, y_column):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x_column, y=y_column, data=data)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title('Scatter Plot')
    plt.show()
```

```python
# Function to create a heatmap
def plot_heatmap(data):
    # Select only numeric columns
    numeric_data = data.select_dtypes(include='number')
    
    # Create a new figure with a specified figure size
    plt.figure(figsize=(10, 8))
    
    # Use seaborn's heatmap function to plot the correlation matrix of the numeric columns in the DataFrame
    sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    
    # Add a title to the plot
    plt.title('Correlation Heatmap')
    
    # Display the plot
    plt.show()
```

```python
# Plot pie chart
plot_pie_chart(data, 'Zone')

```

```python
# Plot scatter plot
plot_scatter_plot(data.head(20), 'Establishment Year', 'Google review rating')
```

```python
numeric_data = data.select_dtypes(include='number')
# Plot heatmap
plot_heatmap(numeric_data)
```
